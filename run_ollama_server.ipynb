```python
import pandas as pd
import numpy as np
# قراءة ملف Excel
df = pd.read_csv('telegram_data.csv')
df.head()
```

```python
import requests
import json

url = "http://localhost:11434/api/chat"  # استبدل هذا بعنوان URL الفعلي للـ API
```

```python
def process_row(row):
  payload = {
  "model": "gemma2:9b-instruct-q5_0",
  "stream": False,
  "messages": [
    {"role": "system", "content": """
    You are an AI assistant specialized in categorizing news related to conflicts and crises. When presented with a news item, your task is to classify it into one of the following categories:

  [الأحداث العسكرية, الإصابات والضحايا, التصريحات السياسية, الوضع الإنساني, التطورات الدولية, التحذيرات والإنذارات]

  Respond only with the appropriate category in Arabic from the list above. Do not provide any additional explanation or commentary. Your response should consist of only one of these categories.
  """},
    { "role": "user", "content": row["text"] }
  ]
  }

  response = requests.post(url, json=payload)
  result = response.json()
  return result['message']['content']
```

```python
# إنشاء DataFrame للتجربة
# يمكنك استبدال هذا بقاعدة البيانات الخاصة بك

# حساب حجم كل جزء
chunk_size = len(df) // 5

# تقسيم DataFrame إلى 5 أجزاء وحفظها في متغيرات منفصلة
df1 = df.iloc[:chunk_size]
df2 = df.iloc[chunk_size:2*chunk_size]
df3 = df.iloc[2*chunk_size:3*chunk_size]
df4 = df.iloc[3*chunk_size:4*chunk_size]
df5 = df.iloc[4*chunk_size:]

# طباعة حجم كل DataFrame للتأكد
print(f"حجم df1: {len(df1)}")
print(f"حجم df2: {len(df2)}")
print(f"حجم df3: {len(df3)}")
print(f"حجم df4: {len(df4)}")
print(f"حجم df5: {len(df5)}")

# طباعة الصفوف الأولى من كل DataFrame للتحقق
print("\nالصفوف الأولى من df1:")
print(df1.head())

print("\nالصفوف الأولى من df2:")
print(df2.head())

print("\nالصفوف الأولى من df3:")
print(df3.head())

print("\nالصفوف الأولى من df4:")
print(df4.head())

print("\nالصفوف الأولى من df5:")
print(df5.head())
```

```python
df1['category'] = df1.apply(process_row, axis=1)
```

```python
df2['category'] = df2.apply(process_row, axis=1)
```

```python
df3['category'] = df3.apply(process_row, axis=1)
```
